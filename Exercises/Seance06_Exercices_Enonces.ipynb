## Exercice 1 - Utilisation de liste des listes
- Voici une liste des fournisseurs chez AdventureWorks ainsi que les prix qu’ils chargent pour les différents produits qu’ils vendent.

| Fournisseur | Chainring Bolts | Chainring Nut | Chainring |
| --- | --- | --- | --- |
| Training Systems | 47,28 | 42,21 | 27,33 |
| Beaumont Bikes | 45,21 | 40,76 | 24,21 |
| Bike Satellite Inc. | 43,21 | 38,56 | 25,11 |

- On vous demande de calculer et d'__afficher la moyenne par produit nommé__.

````python
# on présume que les données sont organisées comme un dictionnaire où on associe une
# liste de prix à chaque fournisseur.
prix = {
    'Training Systems' : [ 47.28, 42.21, 27.33 ],
    'Beaumont Bikes' : [ 45.21, 40.76, 24.21 ],
    'Bike Satellite Inc.' : [ 43.21, 38.56, 25.11 ]
}

# une liste pour les entreprise
noms_produits = ["Chainring Bolts", "Chainring Nut", "Chainring"]

# une liste pour stocker les variables 
prix_produits = []

# pour chaque valeur clé ('fournisseur') dans le dictionnaire 'prix'...
for fournisseur in prix:

    for index, valeur in enumerate(prix[fournisseur]):
            
        if (len(prix_produits) == index):
            prix_produits.append([])

        prix_produits[index].append(valeur)
    
# nous allons traverser les listes de prix pour chaque produit pour calculer la moyenne par produit
# et faire l'affichage par produit nommé.
# 'prix_produits' c'est la liste des listes
# 'valeurs' c'est les listes contenant les prix par produit
for index, valeurs in enumerate(prix_produits):
    print("\nMoyenne par produit {0} : {1}".format(noms_produits[index], sum(valeurs)/len(valeurs)))
````

## Exercice 5 - Analyse par grappes

On veut écrire une fonction qui permet de regrouper des contacts selon la similarité du titre de leur poste. Le résultat devrait permettre de fournir:

- la liste des termes jugés similaires,
- les mots qui sont significatifs dans l'établissement du regroupement,
- la liste des contacts qui sont dans le regroupement.

Votre fonction devrait procéder en 3 étapes:

1. construire la liste de tous les titres de postes disponibles;
2. établir les grappes de titres de postes;
3. associer les noms des contacts à chaque grappe.

Pour construire la structure de grappes, utilisez d'abord un dict. La clé est le premier titre membre de la grappe. La valeur est la liste des titres présents dans la grappe. Une fois toutes les grappes construites, on peut alors n'utiliser qu'une liste de liste, doit la liste des titres constituant les grappes.

````python
#from solutions
# cellule pour exercice 5

from nltk.metrics.distance import jaccard_distance

# Modifier ce seuil pour voir l'impact sur la création des grappes
SEUIL_DISTANCE = 0.4

def cluster_contacts_by_title(first_name, last_name, position_clean):

    # On définit une liste vide qui va contenir tous les titres contenus dans position_clean
    all_titles = []
    
    # Pour chaque liste de listes
    for plist in position_clean:
        
        # Si la liste courante est la liste ne contenant que l'élément '', on passe au suivant
        if plist == ['']:
            continue
        
        # Sinon, on ajoute la liste de titre à la liste globale
        all_titles.extend(plist)

    # Une fois tous les titres ajoutés, s'assure qu'il n'y a pas d'éléments en double
    # On crée un set avec la liste, puis la reconvertit en liste!
    all_titles = list(set(all_titles))

    # On détermine les grappes de titres de poste pour la liste de postes obtenus
    grappes_dict = {}
    for title1 in all_titles:
        
        # On présume que la liste des titres contenus dans la grappe identifiés par title1
        # est initialement vide
        grappes_dict[title1] = []

        # On compare ensuite ce titre avec tous les autres titres
        for title2 in all_titles:
            
            # On ne veut pas comparer deux fois deux titres
            # On sautera donc les titres :
            #    1. qui sont déjà dans une autre grappe
            if title2 in grappes_dict[title1]:
                continue
            #    2. pour lesquels il y a déjà une grappe et que title1 en fait partie
            if title2 in grappes_dict and title1 in grappes_dict[title2]:
                continue

            # On calcule la distance de Jaccard entre les deux titres
            # Attention! Cela peut lever des exceptions, auquel cas on affiche les titres impliqués
            try:
                distance = jaccard_distance(set(title1.split()), set(title2.split()))
            except:
                print(title1.split())
                print(title2.split())
                continue

            # Si la distance entre les deux titres est inférieur au seuil
            # on place ces deux titres dans la même grappe
            
            if distance < SEUIL_DISTANCE:
                grappes_dict[title1].append(title2)
                
    # On peut alors produire la liste des listes de titres associées aux grappes
    # On ne gardera pas les grappes ne contenant qu'un seul titre
    grappes_list = []
    for title in grappes_dict:
        if len(grappes_dict[title]) > 1:
            grappes_list.append(grappes_dict[title])

    # On navigue au travers des grappes et on leur associe les contacts dont les titres sont membres
    # Ici, on utilise un dict dont la clé est un tuple constitué de tous les titres de la grappe!
    # La valeur sera la liste des noms (prénom et nom) des contacts associés à cette grappe
    clustered_contacts = {}
    for cluster in grappes_list:
        # Initialement, aucun contact n'est associé à cette grappe
        clustered_contacts[tuple(cluster)] = []

        # On navigue au travers de tous les contacts
        for i in range(0, len(first_name)):
            
            # On navigue au travers de tous les titres associés à un contact
            for title in position_clean[i]:
                
                # On vérifie que ce titre est présent dans la grappe
                if title in cluster:
                    
                    # Dans ce cas, on ajoute ce contact dans la liste
                    clustered_contacts[tuple(cluster)].append("{0} {1}".format(first_name[i], last_name[i]))
                    
    return clustered_contacts
````

````python
# On détermine toutes les grappes
clustered_contacts = cluster_contacts_by_title(first_name, last_name, position_clean)

# On navigue au travers de toutes les grappes
for titles in clustered_contacts:
    # On affiche les titres composant la grappe
    common_titles_heading = 'Titres en commun: ' + ', '.join(titles)

    # On détermine les termes descriptifs de la grappe en trouvant l'intersection 
    # entre tous les termes de cette grappe
    # On initialise la liste avec les mots du premier titre
    descriptive_terms = set(titles[0].split())
    
    # On ajoute ensuite tous les titres
    for title in titles:
        # On ne préserve que les termes en commun jusqu'à présent
        descriptive_terms.intersection_update(set(title.split()))

    # Si l'ensemble est vide, c'est qu'il n'y a pas de mots en commun dans la grappe
    if len(descriptive_terms) == 0:
        descriptive_terms = ['***Aucun mot en commun***']
    descriptive_terms_heading = 'Termes descriptifs: ' + ', '.join(descriptive_terms)

    print(common_titles_heading)
    print('\n'+descriptive_terms_heading)
    print('-' * 70)
    print('\n'.join(clustered_contacts[titles])) #WHY ISN'T THIS PRINTING PEOPLE'S NAMES???
    print()
````
