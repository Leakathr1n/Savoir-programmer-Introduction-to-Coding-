## Exercice 1
Écrire un script Python qui produit 1000 valeurs aléatoires à l'aide de la fonction `random.choices()`. Cette fonction reçoit en paramètre la liste des valeurs possibles et optionnellement les poids associés à ces valeurs (paramètre `weights`) ainsi que le nombre de valeurs à générer. Le résultat est un tableau contenant `k` éléments.

Utiliser les données et poids suivants:

| Valeur | Poids   |
| ------ | ------  |
| A      | 0,2     |
| B      | 0,4     |
| C      | 0,3     |
| D      | 1,1     |

Insérer les données obtenues dans un pandas et créer un histogramme pour représenter la distribution des quatre valeurs `A`, `B`, `C` et `D`.

```python
# cellule pour exercice 1

# les 1000 valeurs aléatoires

# Définir les valeurs et leurs poids
valeurs = ['A', 'B', 'C', 'D']
poids = [0.2, 0.4, 0.3, 1.1]

# Générer 1000 valeurs aléatoires
k = 1000
resultats = random.choices(valeurs, weights=poids, k=k)

# Afficher les résultats
#print(resultats)

# Dans une DataFrame
resultats_df = pd.DataFrame(resultats, columns=['Valeur'])

print(resultats_df)
```

```python
sns.catplot(kind='count', x='Valeur', data=resultats_df, order =valeurs).set(ylabel='Nombre des fois...')
```

## Exercice 2

- En vous servant du dataset des iris (disponible dans le fichier `iris.csv`), l'un des jeux de données les plus populaires (plus d'infos __[ici](https://en.wikipedia.org/wiki/Iris_flower_data_set)__), afficher un __graphique en nuage de points__ "Scatter plot" (voir la fonction `scatterplot()` dans la __[documentation](https://seaborn.pydata.org/examples/scatterplot_sizes.html)__) permettant de comparer la largeur et la longueur de pétale pour chaque type d'iri recensé.
- Faites ensuite un autre graphique (dans une autre cellule de code!) pour comparer la largeur et la longueur des __[sépales](https://fr.wikipedia.org/wiki/S%C3%A9pale)__.


- __Question pour les fans de la botanique!__ Quelle type d'iris paraît la plus simple à distinguer des autres?

```python
# cellule pour exercice 2

# Load the data into a DataFrame
iris_df = pd.read_csv('iris.csv')

# Display the first few rows of the DataFrame
print(iris_df.head())
```

```python
print('Comparer la largeur et la longueur de pétale pour chaque type d\'iri pétale')
sns.set_theme(style="whitegrid")

g = sns.relplot(
    data=iris_df,
    x="petal.length", y="petal.width",
    hue="variety",
    palette='pastel', sizes=(10, 200),
)

g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
print('Comparer la largeur et la longueur de pétale pour chaque type d\'iri sépales')
sns.set_theme(style="whitegrid")

g = sns.relplot(
    data=iris_df,
    x="sepal.length", y="sepal.width",
    hue="variety",
    palette='pastel', sizes=(10, 200),
)

g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

Pour cet exercice, nous voulons lire le fichier de données `data.txt` avec la fonction `pandas.read_csv()`. **Attention!** le séparateur est le caractère de tabulation `'\t'`.

Une fois le fichier lu, nous afficherons un __graphique en nuage de points__ "Scatter plot" (voir la fonction `scatterplot()` dans la __[documentation](https://seaborn.pydata.org/examples/scatterplot_sizes.html)__) avec la librairie `seaborn`. Le résultat devrait ressembler à l'image suivante :
![graphique](plot.png)
Dans ce graphique :

- la colonne `x` du fichier `data.txt` correspond à l'axe des x;
- la colonne `y` du fichier `data.txt` correspond à l'axe des y;
- le style, la couleur et le type de marqueurs sont spécifiés avec la colonne `bloc` du fichier `data.txt`.


```python
# cellule pour l'exercice 3
data_df = pd.read_csv('data.txt', sep='\t')

print(data_df.head())
```

```python
sns.set_theme(style="whitegrid")

g = sns.relplot(
    data=data_df,
    x="x", y="y",
    hue="bloc",
    palette='pastel', style ='bloc', sizes=(10, 200),
)

g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```
