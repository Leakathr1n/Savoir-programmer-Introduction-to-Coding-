# Question 2: (1pts)

La colonne `Titre`, provenant du DataFrame `films_df`, comporte le titre des films ainsi que l'année de production à la fin de chaque titre avec le format suivant :  
> "Nom du Film *`<Année>`*".

Par exemple :  
> "Le Roi Lion (1994)"

Votre tâche consiste à :
- Déclarer une fonction `obtenir_annee` qui prend en paramètre un titre de film et renvoie l'année de production de ce film sous forme de `int`.
- Appliquer cette fonction à la colonne `Titre` de `films_df` pour créer une nouvelle colonne nommée `Annee`.

```python
# Code Question 2 (fini)
# Montrer la table pour voir comment on peut trancher l'année
films_df.head(1)
```

```python
#Definir la fonctionne 
def obtenir_annee(titre):
    
    # Chercher l'année dans la colonne Titre 
    # L'année est entre des parantheses 
    annee_str = titre.split('(')[-1].split(')')[0]
    
    # Convertir en string
    annee = int(annee_str)
    return annee
    
# Appliquer la fonction à la colonne Titre pour créer une nouvelle colonne Annee
films_df['Annee'] = films_df['Titre'].apply(obtenir_annee)

# Afficher
films_df
```

# Question 3: (1pts)

A partir du DataFrame `films_df`, créez une nouvelle colonne intitulée `Période` pour regrouper les films par générations.

Afin d'y parvenir, vous définirez une fonction `bin_age`. Celle-ci ne prendra qu'un seul paramètre, `year`, et vous devrez l'appliquer sur la colonne `Annee` à l'aide de la méthode `.apply()`.

La fonction `bin_age` aura le comportement suivant:

- Si le film date d'entre 1938 et 1957 (inclusivement), alors la valeur retournée sera "Ancien"
- Si le film date d'entre 1958 et 1977 (inclusivement), alors la valeur retournée sera "Relativement Recent"
- Si le film date de 1978 et plus, alors la valeur retournée sera "Recent"

```python
# Code Question 3 (fini)

year = films_df['Annee']

# Définir la fonction bin_age
def bin_age(year):
    if 1938 <= year <= 1957:
        return "Ancient"
    elif 1958 <= year <= 1977:
        return "Relativement Recent"
    else:
        return "Recent"

# Appliquer la fonction à la colonne Annee pour créer une nouvelle colonne Période
films_df['Période'] = films_df['Annee'].apply(bin_age)

# Afficher la table
films_df.head(15)
```

# Question 5: (1pts)
La colonne `Duree` du DataFrame `film_df` représente la durée du film en minutes. Créez une nouvelle colonne nommée `Duree_str` qui contiendra une chaîne de caractères indiquant la durée du film en heures et en minutes, en respectant précisément le format suivant (sous peine de perdre des points):

>  "*`<heures>`* h *`<minutes>`* mins"
 
si le film dure plus de 60 minutes, sinon
 
> "*`<minutes>`* mins"

**Remarque :**  
La valeur de *`<minutes>`* devra toujours être affichée en utilisant deux chiffres. Si minutes est un nombre à un seul chiffre (c'est-à-dire, de 0 à 9), il devra être précédé d'un 0.

Voir ci-dessous les cinq premières lignes du résultat attendu.

<img src="img/out.png" alt="Drawing" style="width: 650px;"/>

```python
# Code Question 5 (fini)

# Définir une fonction pour formater la durée
def film_length(duree):
    heures = duree // 60
    # Calculer le radical
    minutes = duree % 60
    
    if heures > 60:
        return f"{heures} h {minutes} mins" 
    else:
        return f"{minutes} mins"

#Appliquer la fonctionne
films_df['Duree_str'] = films_df['Duree'].apply(film_length)

# Afficher la table avec la nouvelle colonne
films_df.head(10)
```

# Question 7: (1pts)

Créez une fonction `top_n_films` qui retournera un message (chaine de caractère) contenant les `n` `Titre` de films les plus vues, ainsi que le compte du nombre de fois qu'ils ont étés vus.

`top_n_films` à deux paramètres: 
- `df` (auquel passerez `merged_df` comme argument),
- `n` un paramètre optionnel (avec une valeur de 5 comme argument par defaut) qui peut prendre n'importe quel entier comme argument.

Ci-dessous, une image du résultat attendu (à gauche en utilisant `n` par défaut, à droite en utilisant un `n=10` à titre d'exemple).

Il est à noter que le message doit évoluer dynamiquement pour différentes valeures prise par `n`. 

**Indice:** Regroupez et triez les films dans un premier temps, puis sélectionnez les `n` plus populaires.

```python
# Code Question 7 (fini)

def top_n_films(merged_df, n=5):
    
    # Regrouper par titre de film et compter le nombre de vues
    top_films = merged_df.groupby('Titre')['ID_utilisateur'].count()
    
    # Créer une nouvelle colonne et stocker les données
    top_films = top_films.reset_index(name='Nombre_de_vues')
    
    # Les imagine montre une ordre descendre 
    top_films = top_films.sort_values(by='Nombre_de_vues', ascending=False)
    
    # Sélectionner les n-premiers films
    top_n_films = top_films.head(n)
    
    i_count = -1
    
    # Commencer par le message
    message = f"Les {n} films les plus vus sont :\n-------------------------------- \n"
    
    # Ajouter qqc. à message avec un loop 
    # iterrows(): https://stackoverflow.com/questions/16476924/how-to-iterate-over-rows-in-a-dataframe-in-pandas
    for index, row in top_n_films.iterrows():
        i_count = i_count + 1
        message = message +  "{0}. {1} | {2} vues\n".format(i_count, row['Titre'], row['Nombre_de_vues'])
    
    # La variable que la fonctionne donne
    return message

# Appele la fonction 
message = top_n_films(merged_df, n=5)

print(message)
```

# Question 9: (1.5pts)

À partir du DataFrame `merged_df`:
- Comptez le nombre de vues par `Prenom` et par `Genre`
- Triez les résultats pour chaque `Prenom` en fonction du `Nombre de vues` en ordre décroissant.
- Sélectionnez le septième `Genre` le plus populaire pour chaque `Prenom` (**Indice**: penser à utiliser [nth()](https://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.core.groupby.GroupBy.nth.html) à cette étape)


Affichez le DataFrame final contenant les prénoms des utilisateurs, le septième genre le plus populaire et le nombre de fois que ce genre a été visionné.

```python
# Code Question 9 (fini)

#Compter le nombre de fois où chaque film apparait pour chaque utilisateur.
# Avec .count() on compte le nombre de fois où chaque film apparait
film_count = merged_df.groupby(['Prenom', 'Genre', 'ID_film'])['ID_film'].count()

# Créer une nouvelle colonne et y stocker les variable qui compte l'ID_film
film_count = film_count.reset_index(name='Nombre de vues')

#Trier les résultats pour chaque Prenom en fonction du Nombre de vues en ordre décroissant.
film_count_sorted = film_count.sort_values(by=['Prenom', 'Nombre de vues'], ascending=False)

#Sélectionner le septième Genre le plus populaire pour chaque Prenom
#nth: nous pouvons choisir la 7ème ligne (on commence competer par 0 etc.)
seventh_genre_per_user = film_count_sorted.groupby('Prenom').nth(6).reset_index()

#Afficher le résultat
print(seventh_genre_per_user[['Prenom', 'Genre', 'Nombre de vues']])
```
