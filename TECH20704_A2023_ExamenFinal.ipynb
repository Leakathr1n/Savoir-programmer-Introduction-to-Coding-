```python
import seaborn as sns
import pandas as pd
```

```python
# Question 1 (PAS FINI)
with open("usagers.txt", "r") as text_file:
    file_contents = text_file.readlines()

#créer des listes ou je peux stocker des variables
usager = []
mot_de_passe = []

#iterer et append des listes
for row_data in file_contents:
    columns = row_data.strip().split()
    usager.append(columns[0])
    mot_de_passe.append(columns[1])

#Créer une DataFrame
df = pd.DataFrame({
    'usager': usager,
    'mot_de_passe': mot_de_passe
})

# Verfier 
#J'ai reussi!
# print(df)

#Demander le nom
user_name = input(str("Quel est ton nom d'utilisateur?"))

#demander le mot de passe
mdp_1 = input(str("Mots de passe:"))

#trouver location/index de usager
#C#est le probleme du code; j'ai recu toujours un erreur
index = df[df["usager"]] == [user_name]

#utiliser while true car nous avoins besoin d'un break si l#usager a reussi

while True:

    #comparer le mots de passe avec l'utilisater
    #premiere essaie
    if mdp_1 in index['mot_de_passe']:
       print("Connecté")
        #quitter
        break
    else: 
        print("Usager/Mot de passe invalide")

    #deuxieme essai
    mdp_2 = input("Mots de passe:")
    if mdp_2 in index['mot_de_passe']:
       print("Connecté")
        #quitter
        break
    else: 
        print("Usager/Mot de passe invalide")

    #troisieme essaie
    mdp_3 = input("Mots de passe:")
    if mdp_3 in index['mot_de_passe']:
       print("Connecté")
    else: 
        print("Non-connecté")
else: 
    print("Non-connecté")
```

## Question 2 (12 pts):  

- Chargez les données du fichier `tempete.csv` dans un DataFrame  nommé `df`
- Calculez la moyenne des valeurs de la colonne `MaxWindMPH` et remplacez toutes les valeurs manquantes de cette colonne par la moyenne calculée 
- Remplacez les valeurs manquantes de la colonne `Basin` par la chaîne de caractères `'NA'` 
- Dans colonne `Basin` remplacez la chaîne de caractère `'wp'` par `'WP'` (c'est à dire mettez la chaîne de caractère `'wp'` en majuscule)


```python
# Question 2 (FINI)

#Créer le DataFrame
df =pd.read_csv("tempete.csv")
df.dtypes
#verifier si j'ai reussi
#les valeurs qui n'existe pas, sont declarée avec NaN
#print(df)

#Calculer la moyenne
moyenne = df['MaxWindMPH'].mean()
#verifier si j'ai reussi!
#print(moyenne)

#remplacer les valeurs manquantes de cette colonne par la moyenne calculée
df['MaxWindMPH'] = df['MaxWindMPH'].fillna(moyenne)

#Verifier si j'ai reussi!
#print(df)

# Remplacez les valeurs manquantes de la colonne Basin par la chaîne de caractères 'NA'
df['Basin'] = df['Basin'].fillna('NA')

#Verifier si j'ai reussi
#print(df)

# Dans colonne Basin remplacez la chaîne de caractère 'wp' par 'WP'
df = df.replace('wp', 'WP')
```

## Question 3 (14 pts):

- Créez une nouvelle colonne nommée `Duree` dans le DataFrame `df`. Cette colonne doit représenter la durée, <u>en jours</u>, entre `StartDate` et `EndDate`. Assurez-vous que le résultat soit exprimé en <u>jours entiers</u> (c'est à dire que le dtype de la colonne `Duree` doit etre `int`)
- Créez une nouvelle colonne `Hemisphere` en concaténant les informations des colonnes `Hem_NS`  suivi de `Hem_EW` (ex: si pour une observation donnée, la valeur de `Hem_NS` est `'N'` et la valeur de `Hem_EW`  est `'E'`, alors la valeur de `Hemisphere` sera `'NE'`). Par la suite enlevez (`drop`) les colonnes `Hem_NS`, `Hem_EW` du DataFrame `df` une fois la colonne `Hemisphere` créée.


```python
# Question 3 (FINI)

#Créez une nouvelle colonne nommée Duree dans le DataFrame df
#print(df)
#df.dtypes
#StartDate et EndDate n'est sont pas dans date, alors je dois le convertir avant
df["StartDate"] = pd.to_datetime(df["StartDate"])
df["EndDate"] = pd.to_datetime(df["EndDate"])
#verifier: ils sont en formt date maintenant 
#df.dtypes

#créer une nouvelle colone avec la difference en jours 
df['Duree'] = (df['EndDate'] - df['StartDate']).dt.days

#verifier
#print(df)

#Créez une nouvelle colonne Hemisphere en concaténant les informations des colonnes Hem_NS suivi de Hem_EW 
#(ex: si pour une observation donnée, la valeur de Hem_NS est 'N' et la valeur de Hem_EW est 'E', 
#alors la valeur de Hemisphere sera 'NE'). Par la suite enlevez (drop) les colonnes Hem_NS, 
#Hem_EW du DataFrame df une fois la colonne Hemisphere créée. 

df["Hemisphere"] = df["Hem_NS"] + df["Hem_EW"]

#verifier
#print(df)

#on a besoin de axis = 1 parce que ca dit au Python qu'on veut supprimer une colone
df = df.drop(['Hem_NS', 'Hem_EW'], axis=1)

#verifier
df
```

## Question 5 (12 pts) :

Creez une nouvelle colonne `Ratio` dans le DataFrame `df`, qui sera le ratio de `MaxWindMPH` sur `MinPressure`.  

**Rappel :** ($\text{Ratio} = \frac{\text{MaxWindMPH}}{\text{MinPressure}}$)

Par la suite, créez un DataFrame nommé `df_ratio_sorted` qui contiendra le `Ratio` moyen par `Type` de tempêtes, <u>arrondie à quatre chiffres apres la virgule</u> et <u> ordonné en ordre décroissant selon le `Ratio`</u>  .

Une fois le DataFrame `df_ratio_sorted` obtenu, affichez un message indiquant:
- le `Type` de tempêtes avec le ratio moyen le plus élevé
- le `Type` de tempêtes avec le ratio moyen le moins élevé
- la médiane des ratios moyens (par `Type`)

Le résultat attendu est présenté ci-dessous.

**Remarque** : Pour que les points soient accordés a l'instruction print, l'affichage produit par votre code <u>doit être identique à celui illustrés ci-dessous</u>, et  les valeur des variables demandées <u>ne peuvent pas être *codés en dur* (*'hard coded'*)</u>, c'est-à-dire, vous devez utiliser `.format()` ou une *f-string* pour imprimer la valeur des variables.

<u>**Résultat Attendu:</u>**  

<img src="images/solution.png" alt="Drawing" style="width: 650px;"/>

```python
# Question 5 (FINI)

#Créer une nouvelle colonnne Ratio
df["Ratio"] = df["MaxWindMPH"] / df["MinPressure"]

#verifier
#df

df_selected = df[['Ratio', 'Type']]

#Verifier si j'ai reussi!
#print(df_selected)

#Créer le DataFrame avec la moyenne du Ratio par Type 
df_ratio_sorted = df_selected.groupby('Type').mean().reset_index()


# Seulement 4 decimales 
pd.options.display.float_format = '{:.4f}'.format

# Trier le DataFrame en ordre décroissant par le Ratio
df_ratio_sorted = df_ratio_sorted.sort_values(by='Ratio', ascending=False)

# Verifier
print(df_ratio_sorted)

# Type avec le ratio moyen le plus élevé
max_ratio_row = df_ratio_sorted['Ratio'].iloc[0]
max_ratio_type = df_ratio_sorted['Type'].iloc[0]

#Verifier
#print(max_ratio_row)
#print(max_ratio_type)

# Type avec le ratio moyen le moins élevé
df_ratio_sorted_inverse = df_ratio_sorted.sort_values(by='Ratio', ascending=True)
min_ratio_row = df_ratio_sorted_inverse['Ratio'].iloc[0]
min_ratio_type = df_ratio_sorted_inverse['Type'].iloc[0]

#verifier
#print(min_ratio_row)
#print(min_ratio_type)

# Médiane des ratios moyens 
median_ratio = df_ratio_sorted['Ratio'].median()

#verifier
#print(median_ratio)

print("\n Le type de tempête avec le ratio moyen le plus élevé, le moins élevé, ainsi que \n la médiane des ratios moyens sont les suivants:")
print(f'- Ratio Moyen (par Type) Maximum: {max_ratio_row:.4f} provenant du Type {max_ratio_type}')
print(f'- Ratio Moyen (par Type) Minimum: {min_ratio_row: .4f} provenant du Type {min_ratio_type}')
print(f'- Médiane des ratios moyens (par Type): {median_ratio: .4f}')
```
